```python
# data load and preparation
import pandas as pd
import numpy as np
path_to_file = 'Z://Something/Rabota/Rybachi/Diploma/data_final/Sylv bori declination autumn 2018-2019 all data.xls'
data = pd.read_excel(path_to_file)
data.head()
```

```python
SboriNMF = data['gN, NMF'].dropna().to_numpy()
SboriCMF = data['gN, virt'].dropna().to_numpy()
```

```python
# BetaEstimator import
import os, sys
if os.getcwd() not in sys.path:
    sys.path.append(os.getcwd())
    
from beta_estimation import BetaEstimator

# class object creation
example = BetaEstimator(SboriNMF, SboriCMF, 'S_borin_NMF', 'CMF',
                        effect_mu = np.linspace(0, 150, 25)*np.pi/180, effect_k = np.linspace(-0.9, 5, 20),
                        is_degrees=True, is3d=False)
```

```python
# usage for single sample
example.sample1.rayleigh()
example.sample1.show()
```

```python
# example 2d
example.estimate(n_estim=100, savefile=False)
example.show(savefile=False)
```
